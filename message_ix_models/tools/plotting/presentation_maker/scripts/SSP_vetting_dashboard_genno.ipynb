```python
importlib.reload(ssp_plots)
figs = {}
for region in regions[5:]:
    print(region)
    for sector in all_vars_dict.keys():
        tecs = list(all_vars_dict[sector].values())
        for sub in tecs:
            if len(sub) > 10:
                continue
            try:
                py_df.aggregate_region(sub, append=True)
            except:
                pass
            print(sub)
            figs[ssp_plots.common_starting_substring(sub)] = ssp_plots.plot_ssp_layout(
                py_df, sub, region
            )

    with PdfPages(f"prod_ind_areaplots_1000f_{region}.pdf") as pdf:
        for fig in figs.values():
            pdf.savefig(fig, bbox_inches="tight")
    save_as_svg(f"prod_ind_areaplots_1000f_{region}", figs)
```
